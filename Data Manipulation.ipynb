```python
## 23 genres
```

```python
genre_plot = genres[genres.columns[1:]].sum().plot.pie(rot=0, figsize = (20, 20), title = "Total Number of Games by Genre")
```

```python
fig, ax = plt.subplots(figsize = (12,12));

title = plt.title('Genres by Percentage')
title.set_ha("left")
plt.gca().axis("equal")
pie = plt.pie(genres[genres.columns[1:]].sum(),)
genre_val = genres[genres.columns[1:]].sum()
percents = genre_val.to_numpy() * 100 / genre_val.to_numpy().sum()
labels=[genres.columns]
plt.legend(bbox_to_anchor=(1.50,0.8), loc='upper right',
            labels=['%s, %1.1f %%' % (l, s) for l, s in zip(genres.columns,percents)]);
```
